```python
import rpy2.robjects as robjects
from rpy2.robjects.lib import grdevices
from rpy2.robjects.vectors import IntVector
from IPython.display import Image, display

graphics = robjects.packages.importr('graphics')
```

```python
with grdevices.render_to_bytesio(grdevices.jpeg, width=1024, height=896, res=150) as img:
    graphics.barplot(IntVector((1, 2, 3, 4, 5)), ylab='value')
    
display(Image(data=img.getvalue(), format='jpeg', embed=True))
```

```python
robjects.r('''
        f <- function(r, verbose=FALSE) {
            if (verbose) {
                cat("I am calling f().\n")
            }
            2 * pi * r
        }
        f(3)
        ''')
```

```python
r_f = robjects.globalenv['f']
```

```python
print(r_f.r_repr())
```

```python
r_f = robjects.r['f']
```

```python
r_f
```

```python
res = r_f(3)
```

```python
res
```

```python
# The R 'print' function
rprint = robjects.globalenv.get("print")
stats = importr('stats')
grdevices = importr('grDevices')
base = importr('base')
datasets = importr('datasets')

grid.activate()
```

```python
import math, datetime
import rpy2.robjects.lib.ggplot2 as ggplot2
import rpy2.robjects as ro
from rpy2.robjects.packages import importr
base = importr('base')

mtcars = data(datasets).fetch('mtcars')['mtcars']
```

```python
from rpy2.robjects.packages import importr
grdevices = importr('grDevices')

pp = ggplot2.ggplot(mtcars) + \
     ggplot2.aes_string(x='wt', y='mpg', col='factor(cyl)') + \
     ggplot2.geom_point() + \
     ggplot2.geom_smooth(ggplot2.aes_string(group = 'cyl'),
                         method = 'lm')
pp.plot()
```
